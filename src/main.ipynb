```python
# Load the session
sessionName = "record_p-1_b_2023-12-07_16-43-31"
s = AXCPT.readSession(sessionName)
```

```python
# Get the data for the classifier
cData = s.getClassifierData(metrics="all")
cData
```

```python
# Reformat the data to be input into the classifier

d = cData.copy()

labels = d.loc[:,[("info", "label")]].groupby("epoch").first()["info"]
features = d["features"].unstack(level="channel")
features.columns = ["/".join(s) for s in features.columns.values]
dataset = labels.join(features)
for axis in (0, 1):
    dataset = dataset.dropna(axis=axis, how="all")
dataset.sort_index(axis="columns", inplace=True)

dataset
```
