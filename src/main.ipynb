```python
from src.analysis.processing import AXCPT
import pandas as pd
from src.config import CONFIG
```

```python
# Load all sessions
sessions = AXCPT.readAllSessions(onValueError='raise')
```

```python
sessions
```

```python
# Get the classifier data for each session
cDatas = [s.getClassifierData() for s in sessions]
```

```python
# Combine classifier data into a single dataframe
cData = pd.concat(
    cDatas,
    keys=[s.info["sessionName"] for s in sessions],
    names=["session"]
    )
cData
```

```python
cData["label"].value_counts() / len(cData) * 100
```

```python
import numpy as np

# get rid of hierarchical row labels
cData = cData.reset_index()


# vectroize the participant sesion column
cData['group'] = np.nan
cData['group'] = cData['group'].mask(
                    cData['session'].str.startswith("record_p-1"), 0).mask(\
                    cData['session'].str.startswith("record_p-2"), 1).mask(\
                    cData['session'].str.startswith("record_p-3"), 2).mask(\
                    cData['session'].str.startswith("record_p-4"), 3)
```

```python
from sklearn.model_selection import LeaveOneGroupOut
from sklearn.model_selection import KFold
logo = LeaveOneGroupOut()
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GroupKFold,GridSearchCV,cross_val_score,cross_validate 
from sklearn.model_selection import cross_val_score
```

```python
x = cData["features"]
y = cData['label']
group = cData['group']
```

```python
# SVM
from sklearn.svm import SVC
param_grid = {'classifier__C': [0.1, 1, 10, 100, 1000], 
              'classifier__gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'classifier__kernel': ['rbf']} 

pipe = Pipeline([ ('scaler', StandardScaler()),
                 ('classifier', SVC())])

logo = LeaveOneGroupOut()
gscv=GridSearchCV(pipe,param_grid,cv=logo,n_jobs=16)
gscv.fit(x,y, groups = group )
```

```python
gscv.best_score_ , gscv.best_params_
```

```python
# XGBOOST
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder

# convert the labels into integers
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

param_grid = {
        'classifier__min_child_weight': [1, 5, 10],
        'classifier__gamma': [0.5, 1, 1.5, 2, 5],
        'classifier__subsample': [0.6, 0.8, 1.0],
        'classifier__colsample_bytree': [0.6, 0.8, 1.0],
        'classifier__max_depth': [3, 4, 5]
        }

clf = xgb.XGBClassifier(learning_rate=0.02, n_estimators=600, \
                        objective='multi:softmax',
                    silent=True, nthread=4, num_class = 4)

pipe = Pipeline([ ('scaler', StandardScaler()),
                 ('classifier', clf)])
logo = LeaveOneGroupOut()
gscv=GridSearchCV(pipe,param_grid,cv=logo,n_jobs=16)
gscv.fit(x,y_encoded, groups = group)
```

```python
gscv.best_score_ , gscv.best_params_
```
